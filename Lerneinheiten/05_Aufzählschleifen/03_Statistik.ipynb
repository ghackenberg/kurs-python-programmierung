## 1. Mittelwert

Beachte: Für die Berechnung des Mittelwerts verwenden wir die Funktion `len`.

````python
liste = [0, 7, 9, -4, -10, 3, 95, 5]

mittelwert = 0

for listen_wert in liste:
    mittelwert += listen_wert / len(liste)

print(f"mittelwert = {mittelwert}")
````

## 2. Standardabweichung

Beachte: Für die Berechnung der Standardabweichung verwenden wir die Funktion `pow` sowie die Funktion `sqrt` aus dem Modul `math`.

````python
import math

liste = [0, 7, 9, -4, -10, 3, 95, 5]

# Schritt 1: Mittelwert berechnen

mittelwert = 0

for listen_wert in liste:
    mittelwert += listen_wert / len(liste)

print(f"mittelwert = {mittelwert}")

# Schritt 2: Standardabweichung berechnen

standardabweichung = 0

for listen_wert in liste:
    standardabweichung += pow(listen_wert - mittelwert, 2) / len(liste)

standardabweichung = math.sqrt(standardabweichung)

print(f"standardabweichung = {standardabweichung}")
````

### Implementierung in Python

Beachte: Für die Berechnung des Histogramms verwenden die Funktionen `min` und `max`.

````python
liste = [0, 1, 2, 2, 3, 3.4, 4, 1, 5, 2, 6, 6.7, 7, 8, 9, 9, 8, 7, 7, 10]

bin_anzahl = 11

# Schritt 1: Minimum und Maximum berechnen

minimum = min(liste)
maximum = max(liste)

print(f"minimum = {minimum}")
print(f"maximum = {maximum}")

breite = maximum - minimum

# Schritt 2: Histogramm-Datenstruktur initialisieren

histogramm_mittelwert = []
histogramm_gewichte = []

for bin_index_int in range(bin_anzahl):
    
    bin_mittelwert = minimum + breite * bin_index_int / bin_anzahl + breite / bin_anzahl / 2
    bin_gewicht = 0

    histogramm_mittelwert.append(bin_mittelwert)
    histogramm_gewichte.append(bin_gewicht)

print(f"histogram_mittelwert (vorher) = {histogramm_mittelwert}")
print(f"histogram_anzahl (vorher) = {histogramm_gewichte}")

# Schritt 3: Histogramm berechnen

for listen_wert in liste:

    # Index des Korbes berechnen
    bin_index_float = (listen_wert - minimum) / breite * bin_anzahl
    bin_index_int = min(int(bin_index_float), bin_anzahl - 1) # Spezialfall (das Maximum) erfordert Funktion "min"

    print(f"Zahl {listen_wert} fällt in Korb {bin_index_int} (mit {bin_index_float})")

    # Inhalt des Korbes um eins erhöhen
    histogramm_gewichte[bin_index_int] = histogramm_gewichte[bin_index_int] + 1

print(f"histogram_mittelwert (nachher) = {histogramm_mittelwert}")
print(f"histogram_anzahl (nachher) = {histogramm_gewichte}")
````
